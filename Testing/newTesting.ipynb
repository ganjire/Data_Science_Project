```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image

# Define class index-to-name mapping
class_names = {
    0: "ARMA_1_with_trend",
    1: "ARMA_1_without_trend",
    2: "ARMA_2_with_trend",
    3: "ARMA_2_without_trend",
    4: "ARMA_3_with_trend",
    5: "ARMA_3_without_trend",
    6: "AR_1_with_trend",
    7: "AR_1_without_trend",
    8: "AR_2_with_trend",
    9: "AR_2_without_trend",
    10: "AR_3_with_trend",
    11: "AR_3_without_trend",
    12: "MA_1_with_trend",
    13: "MA_1_without_trend",
    14: "MA_2_with_trend",
    15: "MA_2_without_trend",
    16: "MA_3_with_trend",
    17: "MA_3_without_trend"
}

# Load the trained model
model_path = "final_18.keras"  
model = load_model(model_path)

# Function to preprocess image for CNN
def preprocess_image(image_path, target_size=(224, 224)):
    img = Image.open(image_path).convert("RGB")
    img = img.resize(target_size)  # Resize to match CNN input size
    img_array = np.array(img) / 255.0  # Normalize pixel values
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array, img

# Load the test image
img_path = "testing_images\Bild0.png"  
y_true = "AR_1_without_trend"  # Manually written true label

# Preprocess image
processed_img, img = preprocess_image(img_path)

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()

# Model prediction
y_pred_probs = model.predict(processed_img)
y_pred_class = np.argmax(y_pred_probs, axis=1)[0]  # Get predicted class index
y_pred = class_names[y_pred_class]  # Retrieve class name from dictionary

# Print prediction results
print(f"True Label: {y_true}")
print(f"Predicted Class Index: {y_pred_class}")
print(f"Predicted Label: {y_pred}")

# Check if prediction is correct
if y_true == y_pred:
    print("Prediction is correct!")
else:
    print(" Prediction is incorrect.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from PIL import Image
import os

# Load the CSV file
csv_filename = "Weather Data.csv"  

try:
    df = pd.read_csv(csv_filename)
    print(df.head())  # Display the first few rows
except FileNotFoundError:
    print(f"Error: The file '{csv_filename}' was not found.")
    exit()

# Check if there are enough columns
if df.shape[1] < 2:
    print("Error: The CSV file has fewer than two columns. Ensure the data is correctly formatted.")
    exit()

# Extract the time series from the second column
time_series = df.iloc[:, 1]  

# Normalize the data (optional)
time_series = (time_series - time_series.min()) / (time_series.max() - time_series.min())

# Function to save the time series as an image
def save_timeseries_as_image(time_series, save_path):
    plt.figure(figsize=(5, 5))  # Maintain square format
    plt.plot(time_series, linewidth=2)  # Plot the time series
    plt.axis('off')  # Hide axes
    plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
    plt.close()

# Save the image
image_path = "test_image.png"
save_timeseries_as_image(time_series, image_path)
print(f"Time series saved as: {image_path}")

# Preprocess the image for CNN input
def preprocess_image(image_path, target_size=(224, 224)):
    img = Image.open(image_path).convert("RGB")  # Convert to RGB
    img = img.resize(target_size)  # Resize to match model input size
    img_array = np.array(img) / 255.0  # Normalize pixel values
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

# Process the saved image
processed_image = preprocess_image(image_path)

# Load the trained CNN model
model_path = "final_18.keras"  

if not os.path.exists(model_path):
    raise FileNotFoundError(f"Error: Model file '{model_path}' not found. Ensure the path is correct.")

model = load_model(model_path)

# Make a prediction using the model
predictions = model.predict(processed_image)
predicted_class_index = np.argmax(predictions)  # Get the index of the predicted class

# Define class names (must match the training setup)
class_names = {
    0: "ARMA_1_with_trend",
    1: "ARMA_1_without_trend",
    2: "ARMA_2_with_trend",
    3: "ARMA_2_without_trend",
    4: "ARMA_3_with_trend",
    5: "ARMA_3_without_trend",
    6: "AR_1_with_trend",
    7: "AR_1_without_trend",
    8: "AR_2_with_trend",
    9: "AR_2_without_trend",
    10: "AR_3_with_trend",
    11: "AR_3_without_trend",
    12: "MA_1_with_trend",
    13: "MA_1_without_trend",
    14: "MA_2_with_trend",
    15: "MA_2_without_trend",
    16: "MA_3_with_trend",
    17: "MA_3_without_trend"
}

print(f'csv_filename = "{csv_filename}"') 
# Get the predicted class
predicted_class = class_names.get(predicted_class_index, "Unknown class")
print(f"Predicted class: {predicted_class}")

# Display the image with the prediction as the title
img = Image.open(image_path)
plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.axis("off")  # Hide axes
plt.title(f"Predicted class: {predicted_class}", fontsize=12, color="red")
plt.show()
```
